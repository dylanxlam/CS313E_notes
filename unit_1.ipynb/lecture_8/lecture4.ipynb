```python
import math

class Point (object):
    # constructor
    def __init__ (self, x = 0, y = 0):
        self.x = x # x is the value the user inputs, default (no input) is x = 0
        self.y = y

        # get the distance to another Point object
    def dist (self, other): # other is the other point
        return math.hypot(self.x - other.x, self.y - other.y)

    # string representation of a Point
    def __str__ (self):
        return '(' + str(self.x) + ', ' + str(self.y) + ')'

    # test for equality of two Point Objects
    def __eq__ (self, other):
        tol = 1.0e-6
        return ((abs(self.x - other.x) < tol) and (abs(self.y - other.y) < tol))



class Circle (object):
    # constructor
    def __init__ (self, radius = 1, x = 0, y = 0):
        self.radius = radius
        self.center = Point (x,y)

    # compute the cirumference
    def circumference (self):
        return 2.0 * math.pi * self.radius
    
    # compute the area
    def area (self):
        return math.pi * self.radius * self.radius
    
    # determine if a Point is strictly inside the Circle
    def point_inside (self, p):
        return self.center.dist(p) < self.radius

    # determine if a Circle c is strictly inside the Circle
    def circle_inside (self, c):
        dist_centers = self.center.dist(c.center)
        return dist_centers + c.radius < self.radius
    
    # determine if a Circle is strictly outside this Circle
    def circle_outside (self, c):
        return self.center.dist(c.center) > self.radius
    
    # determine if a Circle overlaps with this Circle
    def circle_overlaps(self, c):
        return not self.center.dist(c.center) < self.radius and self.center.dist(c.center) > self.radius
        return not (self.circle_inside(c)) and not (self.circle_outside(c))
    
    
    def main():
    # create Point Objects
        a = Point()
        b = Point (3,4)
        c = Point (3,4)

        print (a)
        print(b)

        # print the distance between Point ovjects
        print(a.dist(b))
        print(b.dist(a))

        # test for equality
        if b == c: # not equal because they are comparing memory address
            print("Points objects are equal")
        else:
            print("Objects are not equal")


        # test cases for the Circle class

        # create Circle objects
        p = Circle (3, 4, 3)
        q = Circle (1, 3, 2)
        r = Circle (2, 5, 4)

        # circumference of p
        print(p.circumference())

        # print the area of q
        print(q.area())

        # is b inside of r
        if (r.point_inside(b)):
            print("Point is inside the Circle")
        else:
            print("Point is not inside the Circle")

    main()
```
